```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
data = make_classification(n_classes=2, n_samples=10000, n_features=20)
```

```python
X = data[0]
y = data[1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtr, Xte, ytr, yte = train_test_split(X, y, train_size=0.8, random_state=14)
```

```python
model.fit(Xtr, ytr)
```

```python
y_pred = model.predict(Xte)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(yte, y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(yte, y_pred))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
y_pred_prob = model.predict_proba(Xte)
```

```python
fpr, tpr, threshold = roc_curve(yte, y_pred_prob[:,1])
```

```python
auc_score = roc_auc_score(yte, y_pred_prob[:,1])
```

```python
auc_score
```

```python
plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], "g--")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC-AUC")
plt.show()
```
