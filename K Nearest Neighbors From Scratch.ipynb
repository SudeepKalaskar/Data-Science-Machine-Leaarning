```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = make_circles(n_samples=40, factor=0.5)
X = np.round(data[0], decimals=2)
y = np.round(data[1], decimals=2)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], label="class 0")
plt.scatter(X[y==1, 0], X[y==1, 1], label="class 1")
plt.legend()
plt.xlabel("X Values")
plt.ylabel("y Values")
plt.title("Circle Data")
plt.show()
```

# This data is called Non Linear Separable Data

```python
df = pd.DataFrame({"Col1":X[:,0], "Col2":X[:,1], "Target":y})
df
```

```python
new_point = np.array([0.5, -0.71])
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], label="class 0")
plt.scatter(X[y==1, 0], X[y==1, 1], label="class 1")
plt.scatter(new_point[0], new_point[1], label="New Point")
plt.legend()
plt.xlabel("X Values")
plt.ylabel("y Values")
plt.title("Circle Data")
plt.show()
```

```python
dist = np.sqrt(np.sum(np.square(X[0]-new_point[0]) + np.square(X[1]-new_point[1])))
```

```python
dist
```

```python
k=3
```

```python
df["Dist"] = dist
```

```python
df.head()
```

```python
df.sort_values("Dist").head(k)
```
