```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = make_circles(n_samples=40, factor=0.5)
X = np.round(data[0], decimals=2)
y = np.round(data[1], decimals=2)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], label="class 0")
plt.scatter(X[y==1, 0], X[y==1, 1], label="class 1")
plt.legend()
plt.xlabel("X Values")
plt.ylabel("y Values")
plt.title("Circle Data")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc = KNeighborsClassifier(n_neighbors=3)
```

```python
knc.fit(X, y)
```

```python
new_point = np.array([0.5, -0.71])
```

```python
knc.predict([new_point])
```

```python
from sklearn.datasets import make_regression
```

```python
reg = make_regression(n_samples=40, n_features=1, noise=10)
```

```python
X = reg[0]
y = reg[1]
```

```python
plt.scatter(X, y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor(n_neighbors=5)
```

```python
knr.fit(X, y)
```

```python
new_point = np.array([[0.5]])
```

```python
pred = knr.predict(new_point)
pred
```

```python
plt.scatter(X, y)
plt.scatter(new_point, pred)
```
