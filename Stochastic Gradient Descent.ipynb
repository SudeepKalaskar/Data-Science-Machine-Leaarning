```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate sample data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X.ravel() + np.random.randn(100)
```

```python
# Parameters for gradient descent
lr = 0.05  # learning rate
b = -43  # intercept
m = 96  # slope or coefficient of x
epochs = 30
```

```python
# Stochastic gradient descent parameters
sgd_lr = 0.05  # learning rate for stochastic gradient descent
```

```python
for epoch in range(epochs):
    for i in range(len(X)):
        # Stochastic gradient descent
        X_i = X[i].ravel()
        y_i = y[i]

        b_slope = -2 * (y_i - m * X_i - b)
        m_slope = -2 * (y_i - m * X_i - b) * X_i

        b -= sgd_lr * b_slope
        m -= sgd_lr * m_slope

    y_pred = m * X.ravel() + b

    # Plot the results
    plt.scatter(X, y)
    plt.plot(X, y_pred, "g-")
    plt.title(f"Epoch {epoch+1}")
    plt.xlabel("X")
    plt.ylabel("y")
plt.show()
```
