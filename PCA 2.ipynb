```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA
```

```python
x,y = make_classification(n_features=5, n_classes=2, n_samples=20)
```

```python
x
```

```python
x1_mean = x[:,0].mean()
x2_mean = x[:,1].mean()
x3_mean = x[:,2].mean()
x4_mean = x[:,3].mean()
x5_mean = x[:,4].mean()
```

```python
x1 = x[:,0]-x1_mean
x2 = x[:,1]-x2_mean
x3 = x[:,2]-x3_mean
x4 = x[:,3]-x4_mean
x5 = x[:,4]-x5_mean
```

```python
data = np.array([x1, x2, x3, x4, x5]).T
data
```

```python
cov_matrix = np.cov(data.T)
cov_matrix
```

```python
eig_value, eig_vec = np.linalg.eig(cov_matrix)
```

```python
eig_value
```

```python
eig_vec
```

```python
pc = np.dot(data, eig_vec)
pc
```

```python
principal_comp = pc[:,:2]
principal_comp
```
